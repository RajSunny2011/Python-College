```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder

dat = {
    'color':['red','green','blue','red'],
    'size':[10,15,20,12],
    'price':[20,30,40,25]
}
df = pd.DataFrame(dat)
df
```

```python
# feature scaling
# z-score normalization 5  
scaler = StandardScaler()
df[['size','price']] = scaler.fit_transform(df[['size','price']])
df
```

```python
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df[['color']]).toarray()
encoded_df = pd.DataFrame(data=encoded_data, columns=encoder.get_feature_names_out(['color']))
encoded_df
```

```python
final_df = pd.concat([df.drop('color',axis=1), encoded_df],axis=1)
final_df
```
