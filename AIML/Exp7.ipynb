```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
np.random.seed(42)

feature_1_class_0 = np.random.normal(2, 1, 100)
# Generates 100 valuees from a normal distribution centered at 2, with a standard deviation of 1
feature_2_class_0 = np.random.normal(3, 1, 100)
# Generates 100 valuees from a normal distribution centered at 3, with a standard deviation of 1

feature_1_class_1 = np.random.normal(6, 1, 100)
# Generates 100 valuees from a normal distribution centered at 6, with a standard deviation of 1
feature_2_class_1 = np.random.normal(5, 1, 100)
# Generates 100 valuees from a normal distribution centered at 6, with a standard deviation of 1
```

```python
print(f"feature 1 class 0:\n{feature_1_class_0}")
```

```python
print(f"feature 2 class 0:\n{feature_2_class_0}")
```

```python
print(f"feature 1 class 1:\n{feature_1_class_1}")
```

```python
print(f"feature 2 class 1:\n{feature_2_class_1}")
```

```python
feature_class_0 = np.column_stack((feature_1_class_0,feature_2_class_0))
feature_class_1 = np.column_stack((feature_1_class_1,feature_2_class_1))
labels_class_0 = np.zeros(100)
labele_class_1 = np.ones(100)
features = np.vstack((feature_class_0, feature_class_1))
labels = np.concatenate((labels_class_0, labele_class_1))

data = pd.DataFrame(features, columns = ["Feature_1", "Feature_2"])
data["Target"] = labels

x = data[["Feature_1", "Feature_2"]]
y = data["Target"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier(max_depth=3, random_state=42)

clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)

accuracy  = metrics.accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")

plt.figure(figsize=(13,5))
plot_tree(clf, feature_names=["Feature 1", "Feature 2"], class_names=["Class 0", "Class 1"], filled=True)
plt.title("Decision Tree Visulization")
plt.show()
```
