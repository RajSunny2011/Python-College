```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('supermarket_sales.csv')
df.info()
```

```python
plt.figure()
plt.plot(df['Unit price'],df['Total'],marker = '*')
plt.title('Unit price')
plt.xlabel('Total')
plt.ylabel('cogs')
plt.show()
```

```python
plt.figure()
plt.bar(df['Quantity'],df['cogs'])
plt.title('Bar Chart')
plt.xlabel('Quantity')
plt.ylabel('cogs')
plt.show()
```

```python
plt.figure()
sns.histplot(df['Quantity'],bins = 6)
plt.title('Histogram')
plt.show()
```

```python
plt.figure()
sns.boxplot(x = 'Quantity', y = 'Tax 5%', data=df)
plt.title('Boxplot')
plt.show()
```

```python
af = df[['Unit price','Quantity','Tax 5%','cogs','gross income','Rating']]
corr = af.corr()
sns.heatmap(corr, annot=True, cmap="crest")
plt.show()
```
